# Question : 3
 Calculate the average attendance of students in the "Engineering department"

```python
#Answer to the question No:3
#Calculate the average attendance of students in the "Engineering department"
engineering_students = df[df["department"] == "Computer Science"]
average_attendance = engineering_students["attendance"].mean()
print(f"Average attendance of engineering students: {average_attendance}")
```

# Question Number:4
. Replace missing values in "study_abroad_country" with "Not Available".


```python
# Answer to the question No:4
# Replace missing values in "study_abroad_country" with "Not Available".
print("Missing Data:\n",df.isnull())
missing_value= df.fillna({"study_abroad_country": "Not Available"})

print(f"Missing values in 'study_abroad_country' have been replaced with 'Not Available'.{missing_value}")
```

# Question Number:5
. Generate a line plot to show the trend in CGPA over passing years.

```python
# Answer to the question No:5
#Generate a line plot to show the trend in CGPA over passing years.

import pandas as pd
import matplotlib.pyplot as plt

cgpa_trend = df.groupby('graduation_date')['cgpa'].mean()
plt.figure(figsize=(10, 6))
plt.plot(cgpa_trend.index, cgpa_trend.values, marker='o', linestyle='-', color='r')
plt.title('Trend in CGPA Over Passing Years')
plt.xlabel('Graduation Year')
plt.ylabel('Average CGPA')
plt.grid(True)
plt.show()
```
